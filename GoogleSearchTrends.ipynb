# Extracting the google search trends results using pytrends package 

```python
import pytrends
import pandas as pd
from pytrends.request import TrendReq

 

# Connect to Google using a Proxie due to the Google Rate Limit
pytrends = TrendReq(hl='en-US', tz=300)

 

# Create Keyword_list
kw_list = ['Gap Inc.']



# Records Interest Over Week from timeframe Period Above for the entire United States
pytrends.build_payload(kw_list, cat=0, timeframe='2015-01-01 2020-01-1', geo='US', gprop='')

 

# Retrieve the interest over time data
interest_over_time = pytrends.interest_over_time()
interest_over_time.drop(interest_over_time[['isPartial']], axis=1, inplace=True)

 

# Rename columns if needed
interest_over_time.rename(columns={kw_list[0]: 'Interest'}, inplace=True)

 

# Save Dataframe from Pandas to CSV
interest_over_time.to_csv('/Users/avni/Desktop/gap.csv', sep=',', na_rep='0')
```
